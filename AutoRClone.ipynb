````python
#@title **Rename your Downloaded Service Account `JSONs` to numbers**

#@markdown > *Once you have downloaded your Service Account JSON files to Google Colab, it is best you rename them to numbers like `1.json, 2.json, 3.json` and so on. Running this cell is not required for AutoRClone though.*

# CD to AutoRClone Folder and print text
!cd /content/AutoRclone
print("||----- Renaming JSON Files... -----||")

# Rename.py Script
import os
base = "accounts"
count = 1
for file in os.listdir(base):
    abspath = os.path.join(base,file)
    destpath = os.path.join(base,f"{count}.json")
    os.rename(abspath,destpath)
    count += 1

# List files in accounts folder
!ls /content/AutoRclone/accounts
print("||----- All files renamed! -----||")
````
